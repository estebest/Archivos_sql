```python
import pandas as pd
import numpy as np
```

```python
df_links = pd.read_excel('links_google.xlsx')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Configura el navegador en modo oculto
options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver = webdriver.Chrome(options=options)

textos_extraidos = []

# Iterar con contador
for i, url in enumerate(df_links['link'], start=1):
    print(f"Fila {i} de {len(df_links)}: ", end='')

    if not url or pd.isna(url) or url.strip() == '':
        print("sin URL — se omite.")
        textos_extraidos.append(None)
        continue

    try:
        print(f"Visitando: {url}")
        driver.get(url)
        time.sleep(3)

        try:
            plan = driver.find_element(By.LINK_TEXT, 'Plan de estudios')
            plan.click()
            time.sleep(3)
        except:
            print("No se encontró 'Plan de estudios' — usando página principal")

        texto = driver.find_element(By.TAG_NAME, 'body').text
        textos_extraidos.append(texto)

    except Exception as e:
        print(f"Error al procesar la página: {e}")
        textos_extraidos.append(None)

driver.quit()

# Asignar resultados solo a las filas correspondientes
df_links= df_links.copy()  # Evita SettingWithCopyWarning
df_links['texto_plan_estudios'] = textos_extraidos
```

```python
while len(textos_extraidos) < len(df_links):
    textos_extraidos.append(pd.NA)

df_links['texto_plan_estudios'] = textos_extraidos

```
